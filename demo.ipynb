
Silhouette Score: <br>
The Silhouette Score measures how similar an object is to its own cluster compared to other clusters. It ranges from -1 to 1:<br>

+1: The sample is far away from the neighboring clusters. <br>
0: The sample is on or very close to the decision boundary between two neighboring clusters.<br>
-1: The sample is assigned to the wrong cluster.<br>


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate example data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Fit KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Calculate silhouette score
sil_score = silhouette_score(X, labels)
print(f"Silhouette Score: {round(sil_score, 2)}")

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
plt.title(f'Silhouette Score: {round(sil_score, 2)}')
plt.show()
```

Davies-Bouldin Index: <br>
The Davies-Bouldin Index measures the average similarity ratio of each cluster with its most similar cluster. <br>
It is defined as the average ratio of within-cluster distances to between-cluster distances:<br>

Lower values: Indicate better clustering performance.

```python
from sklearn.metrics import davies_bouldin_score

# Calculate Davies-Bouldin Index
db_index = davies_bouldin_score(X, labels)
print(f"Davies-Bouldin Index: {round(db_index,2)}")

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
plt.title(f'Davies-Bouldin Index: {round(db_index, 2)}')
plt.show()
```

### Generate another dataset with more ambiguous clusters centroids
